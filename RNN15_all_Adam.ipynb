```python
### Loading Model ###
%matplotlib inline
import pylab
import cPickle
import scipy
import random
import matplotlib.gridspec as gridspec
import numpy
import matplotlib.pyplot as plt
import matplotlib.lines as lin
import matplotlib
import h5py
import theano
import theano.tensor as T
matplotlib.rcParams['savefig.dpi'] = 2 * matplotlib.rcParams['savefig.dpi']
f = file('nat_parasols_sigUpdate_lam_0_nk_1_convRNN_15_poiss_alln_adam_undone.save','rb')
#f = file('nat_parasols_sigUpdate_lam_0_nk_1_convRNNdir_DoG_2_15_poiss2.save_dir_DoG_2_dir_DoG_2','rb')
params = []
params.append(cPickle.load(f))

```

```python
## Relative Information Measure
normalizedLL = numpy.zeros((9,))
for i_n in xrange(0,9):
    Eyx = params[0][9][i_n,:]/49.
    Eyx = T.cast(Eyx,'float64')
    true_y = params[0][10][i_n,:]

    # LL of p(s|x)
    LL = (true_y*T.log2(Eyx))+T.log2(T.exp(-Eyx*49.))#-T.log2(scipy.misc.factorial(true_y))
    #2**LL.eval()

    # LL of p(s)
    meanFR = numpy.array(sum(true_y)/3600./49,dtype='float64')
    LL_meanFR = (true_y*T.log2(meanFR))+T.log2(T.exp(-meanFR*49.))#-T.log2(scipy.misc.factorial(true_y))

    # Put them together
    normalizedLL[i_n] = (T.sum(LL - LL_meanFR).eval()*120.)/3600.

normalizedLL
```

```python
## Load TestResponses to look at relative information across trials
dataset='../TestResponses.mat'
f = h5py.File(dataset)
TestResponses = numpy.array(f['TestResponses'])
```

```python
## Look at relative information across trials
normLL_sum = numpy.zeros((9,))
normLL_mean = numpy.zeros((9,))
normLL_std = numpy.zeros((9,))
for i_n in xrange(0,9):
    print i_n
    ntrials = 59;
    testMovie_len = 3600;
    raster= numpy.zeros((3600,59))
    for i in xrange(0,ntrials):
       raster[:,i] = TestResponses[xrange(i*testMovie_len,(i+1)*testMovie_len),i_n]


    normalizedLL_ind = numpy.zeros((59,))
    for i_trial in xrange(10,59):
        Eyx = params[0][9][i_n,:]/49.
        Eyx = T.cast(Eyx,'float64')
        true_y = raster[:,i_trial]
        all_y = params[0][10][i_n,:]
        # LL of p(s|x)
        LL = (true_y*T.log2(Eyx))+T.log2(T.exp(-Eyx))#-T.log2(scipy.misc.factorial(true_y))
        #2**LL.eval()

        # LL of p(s)
        meanFR = numpy.array(sum(all_y)/3600./49,dtype='float64')
        LL_meanFR = (true_y*T.log2(meanFR))+T.log2(T.exp(-meanFR))#-T.log2(scipy.misc.factorial(true_y))

        # Put them together
        normalizedLL_ind[i_trial] = (T.sum(LL - LL_meanFR).eval()*120.)/3600.
    normLL_sum[i_n] = numpy.sum(normalizedLL_ind[10:59])
    normLL_mean[i_n] = numpy.mean(normalizedLL_ind[10:59])
    normLL_std[i_n] = numpy.std(normalizedLL_ind[10:59])
```
